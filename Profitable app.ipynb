To decide for this, we use, 

- a dictionary to keep track of the **`reviews`** of the highest frequency as a criterion

- We keep track of the records that are updated the last time. In this regard, the record with the highest review should be the one of our main focus for the latest record.

```python
max_review = {}

for row in google_data_list[1:]:
    app_name = row[0]
    app_review = float(row[3])
    
    if app_name in max_review and max_review[app_name] < app_review:
        max_review[app_name] = app_review
        
    elif app_name not in max_review:
        max_review[app_name] = app_review
        
```

```python
unique_google_list = []
duplicate_google_list = []

for row in google_data_list[1:]:
    name = row[0]
    total_app_reviews = float(row[3])
    
    if (max_review[name] == total_app_reviews) and row not in duplicate_google_list:
        unique_google_list.append(row)
    else:
        duplicate_google_list.append(row)

        
print(unique_google_list[:5])
```

Here ™ and 😜, emoji or special charater are supposed 
to make our function show False result even for some English inputs, hence, 
we shall be ignoring up to 3 special characters 
that have ascii values more than 127.

```python
def is_english(strings):
    count = 0

    for character in strings:
        
        if ord(character) > 127 and count > 3:
            return False
        
        if ord(character) > 127:
            count += 1
        
    return True
```

- Frequency table in ascending order

- make the function usable for all columns

- represent the values in percentage


```python
def freq_table(dataset, index):
    table = {}
#     table for frequency
    total = 0
#     total app
    
    for row in dataset:
        total += 1
        column_value = row[index]
        
        if column_value in table:
            table[column_value] += 1
            
        else:
            table[column_value] = 1
            
        table_percentages = {}
        
    for key in table:
        percentage = (table[key] / total) * 100
        table_percentages[key] = percentage

    return table_percentages
```

## sorting

```python
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
#         appending tuple, from dictionary
        
    table_sorted = sorted(table_display, reverse = True)
    
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

We are going to calculate the averate number of installs for google play apps, but, for ios apps, there is no column as such hence, we are going to use rating_count_tot column to determine the average numbers of user ratings per app genre on the app store.

```python
genres_ios = freq_table(free_apple_final_dataset, -5)
# prime_genre column

for genre in genres_ios:
    total = 0
#     to count the numbers of ratings per genre

    len_genre = 0
    
    for app in free_apple_final_dataset:
        genre_app = app[-5]
        
        if genre_app == genre:
            n_ratings = float(app[5])
            total += n_ratings
            len_genre += 1
            
    avg_n_ratings = total / len_genre
    print(genre, ':', avg_n_ratings)
```

We can see that most values are open-ended(100+, 1000+ etc)

---

For instance, we don't know whether an app with 100,000+ installs has 100,000 installs, 200,000, or 350,000. However, we don't need very precise data for our purposes — we only want to find out which app genres attract the most users.

We're going to leave the numbers as they are, which means that we'll consider that an app with 100,000+ installs has 100,000 installs, and an app with 1,000,000+ installs has 1,000,000 installs, and so on. To perform computations, however, we'll need to convert each install number from a string to a float. This means we need to remove the commas and the plus characters, or the conversion will fail and cause an error.

```python
android_category = freq_table(free_google_final_dataset, 1)

android_category
```

```python
for category in android_category:
    total = 0
#     sum of installs specific to each genre

    len_category = 0
#     length of numbers of apps specific to each genre

    for row in free_google_final_dataset:
        category_app = row[1]
#         category column

        if category_app == category:
            number_of_installs = row[5]
#             installs column

            number_of_installs = number_of_installs.replace('+', '')
            number_of_installs = number_of_installs.replace(',', '')
            
            total += float(number_of_installs)
            len_category += 1
            
    avg_n_installs = total / len_category
    print(category, ':', avg_n_installs)
```

From the above findings, 

- `Communication` is the highest installed apps for the user preference

- `Entertainment` and `Game` Catergory comes there after

Hence the trend is more in these categories for people to have more install records. Hence, our free app should be more installed reaching out to the market should we build an app related to these area
